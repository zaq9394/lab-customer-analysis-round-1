```python
import pandas as pd  
import numpy as np
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
display(file1.head())
display(file2.head())
display(file3.head())
```

```python
file1=file1.rename(columns={'ST':'State','GENDER':'Gender'})
file2=file2.rename(columns={'ST':'State','GENDER':'Gender'})
file3.columns
file3=file3[['Customer', 'State','Gender', 'Education', 'Customer Lifetime Value', 
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
file1=file1[['Customer', 'State','Gender', 'Education', 'Customer Lifetime Value', 
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
file2=file2[['Customer', 'State','Gender', 'Education', 'Customer Lifetime Value', 
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
display(file3.head())
display(file2.head())
display(file1.head())
```

```python
data=pd.concat([file1,file2,file3],axis=0)
data.shape
data.head()
```

```python
data.dtypes
data.info()
```

```python
#Income, Monthly Premium Auto and Total Claim Amount are numerical and the rest of are categorical.
```

```python
data=data.drop(['Education','Number of Open Complaints'],axis=1)
data.head()
```

```python
#data['Customer Lifetime Value']=list(map(lambda x: str(x).replace('%', ''),data['Customer Lifetime Value']))
col=[]
for i in data['Customer Lifetime Value']:
    
        col.append(str(i).replace('%',''))
data['Customer Lifetime Value']=col


```

```python
data.head()
```

```python
data['Customer Lifetime Value'] =  pd.to_numeric(data['Customer Lifetime Value'], errors='coerce')
data = data.drop_duplicates()
data = data.reset_index(drop=True)
data.head()
```

```python
data.isna().sum()
mean_median_home_value = data['Customer Lifetime Value'].mean()
data['Customer Lifetime Value'] = data['Customer Lifetime Value'].fillna(mean_median_home_value)
data['Gender'].value_counts(dropna=False)
data['Gender']=data['Gender'].fillna('F')
data.head()
```

```python
data['Customer Lifetime Value']=list(map(lambda x: x/100,data['Customer Lifetime Value']))
data.head()
```
